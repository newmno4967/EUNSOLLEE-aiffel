```python
def sol():
    #     
    file_path = "/content/drive/MyDrive/20./Avengers.txt"
    with open(file_path, "r") as file:
        text = file.read()

    #  
    import re                              #    
    text = text.replace("\n", " ")         #   
    text = re.sub(r'[^\w\s]', ' ', text)   #   
    text = " ".join(text.split()).lower()  #       

    #   count
    text_count = {}                 #   
    for count in text.split():      #       
        if count in text_count:     #    
            text_count[count] += 1  #   1 
        else:                       #   
            text_count[count] = 1   #      1 

    #  sorting
    sorted_text_count = dict(sorted(text_count.items()))                                                   #  key  (abc)
    sorted_text_count = dict(sorted(sorted_text_count.items(), key=lambda item: item[1], reverse = True))  #   ()
    num_text = {key: i for i, key in enumerate(sorted_text_count.keys())}                                  #      

    #  
    user = input("  : ")         #    
    if user in num_text:                          #      
        return num_text[user]                     #     
    else:                                         #      
        return '  !'                #    
```

```python
#    
sol()
```
